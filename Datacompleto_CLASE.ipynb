```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('Ventas_totales.csv')
data
```

```python
#Identificar valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
#Reemplazaremos valores nulos del dataframe con 'bfill'
data1 = data.fillna(method='bfill')
data1
```

```python
#Identificar valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
data2=data.copy()
data2=data2.fillna(round(data.mean(),1))
data2
```

```python
#Identificar valores nulos
valores_nulos=data1.isnull().sum()
valores_nulos
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize=(15,8))
data2.plot(kind='box', vert=False)
plt.title('Valores atípicos del Dataframe')
plt.show() #dibujamos el diagrama
```

```python
#Método aplicando desviación estandar. Encuentro valores extremos
y= data2
Límite_Superior = y.mean() + 3*y.std()
Límite_Inferior = y.mean() - 3*y.std()
print('Límite superior permitido', Límite_Superior)
print('Límite inferior permitido', Límite_Inferior)
```

```python
#Encontrar outliers del dataframe
outliers= data2[(y > Límite_Superior)|(y < Límite_Inferior)]
outliers
```

```python
#Obtenemos datos sin outliers del dataframe
data3= data2[(y <= Límite_Superior)&(y >= Límite_Inferior)]
data3
```

```python
#Eliminamos columnas innecesarias
data4=data3.drop(["indice_tiempo"],axis=1)
data4
```

```python
#Remplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data4.copy()
data_clean=data_clean.fillna(round(data4.mean(),1))
data_clean
```

```python
#Identificación de valores nuloss
valoresnull=data_clean.isnull().sum()
valoresnull
```

```python
#PROCEDIMIENTOS CUANTILES PARA ELIMINAR OUTLIERS EN DATAFRAME

#Método aplicando Cuartiles.Encuentro cuartiles 0.25 y 0.75

y=data2

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 + 1.5*iqr
print ("Limite superior premitido", Limite_Superior_iqr) 
print ("Limite Inferior premitido", Limite_Inferior_iqr) 
```

```python
outliers_iqr=data2[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#
data4_iqr=outliers_iqr.drop(["indice_tiempo"], axis=1)
data4_iqr
```

```python
#Remplazamos valores atipicos (nulos) del dataframe
#Realizamos una copia del dataframe
data5_iqr=data4_iqr.copy()
data5_iqr=data5_iqr.fillna(round(data4_iqr.mean(),1))
data5_iqr
```

```python
#Unimos la comlumna "Indice tiempo" en el dataframe
indice_tiempo=data2["indice_tiempo"]
Datos_limpios = pd.concat([indice_tiempo, data5_iqr], axis=1)
Datos_limpios
```

```python
#Identificación de valores nuloss
valoresnull=Datos_limpios.isnull().sum()
valoresnull
```

```python
#Converir a CSV
#ARCHIVOS DE VARIABLE 
Datos_limpios.to_csv("Datos_limpios.csv")
```
