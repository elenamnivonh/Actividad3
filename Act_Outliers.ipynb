```python
#Importamos las librerias correspondientes a esta actividad
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga de archivo CSV
dataframe= pd.read_csv("Ventas_totales.csv")
dataframe
```

```python
#Identificación de valores nulos 
valoresnull=dataframe.isnull().sum()
valoresnull
```

```python
#Realizamos una copia del dataframe
data2=dataframe.copy()
```

```python
#El primer cambio corresponde a la sustitución de valores nulos en las variables 
#Salon_ventas
data2["salon_ventas"]=data2["salon_ventas"].fillna(round(dataframe["salon_ventas"].mean(),1))
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustitución de valores nulos con promedio en la variable subtotal_ventas_alimentos_bebidas
data2["subtotal_ventas_alimentos_bebidas"]=data2["subtotal_ventas_alimentos_bebidas"].fillna(round(dataframe["subtotal_ventas_alimentos_bebidas"].mean(),1))
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustitución de valores en variable otros
data2["otros"]=data2["otros"].fillna(round(dataframe["otros"].mean(),1))
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustitución de valores nulos en:
#tarjetas_credito
# almacen
#indumentaria_calzado_textiles_hogar  
#electronicos_articulos_hogar     
data2["tarjetas_credito"]=data2["tarjetas_credito"].fillna(round(dataframe["tarjetas_credito"].median(),1))
data2

data2["almacen"]=data2["almacen"].fillna(round(dataframe["almacen"].median(),1))
data2

data2["indumentaria_calzado_textiles_hogar"]=data2["indumentaria_calzado_textiles_hogar"].fillna(round(dataframe["indumentaria_calzado_textiles_hogar"].median(),1))
data2

data2["electronicos_articulos_hogar"]=data2["electronicos_articulos_hogar"].fillna(round(dataframe["electronicos_articulos_hogar"].median(),1))
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustitución de valores nulos con mediana en la variable tarjetas de debito
data2["tarjetas_debito"]=data2["tarjetas_debito"].fillna(round(dataframe["tarjetas_debito"].median(),1))
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustituir valores nulos por un valor en concreto en la variable otros medios 
#Variable Otros_medios
data2["otros_medios"]= data2["otros_medios"].fillna(3450700.50)
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustituir valores nulos por un valor en concreto en la variable otros medios 
#Variable Otros
data2["otros"]= data2["otros"].fillna(5776003.532)
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  ("ffill")
#Bebidas
data2["bebidas"]=data2["bebidas"].fillna(method="ffill")
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  ("ffill")
#Panaderia
data2["panaderia"]=data2["panaderia"].fillna(method="ffill")
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás "backward fill" ("bfill")
#Lacteos
data2["lacteos"] =data2["lacteos"].fillna(method="bfill")
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustituir valores nulos por valores no nulos hacia adelante  ("ffill")
#Carnes
data2["carnes"]=data2["carnes"].fillna(method="ffill")
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás  ("bfill")
#verduleria_fruteria
data2["verduleria_fruteria"] =data2["verduleria_fruteria"].fillna(method="bfill") 
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Sustituir valores nulos por valores no nulos hacia atrás  ("bfill")
#alimentos_preparados_rostiseria
data2["alimentos_preparados_rotiseria"] =data2["alimentos_preparados_rotiseria"].fillna(method="bfill")
data2
```

```python
#Identificación de valores nuloss
valoresnull=data2.isnull().sum()
valoresnull
```

```python
#Convertir DataFrame a CSV
data2.to_csv("VentasSNA.csv")
```

```python
#Valores extremos de límite superior y límite inferior.
#Primera variable
x1=data3["ventas_precios_corrientes"]
LS1= x1.mean() + 3*x1.std()
LI1= x1.mean() - 3*x1.std()
print("Limite superior permitido", LS1)
print("Limite inferior permitido", LI1)
```

```python
#Outliers de x1
outliersx1= data3[(x1>LS1)|(x1<LI1)]
outliersx1
```
